Q.6 You should use a Z-test rather than a t-test under the following conditions:

1. Population Standard Deviation (œÉ) is Known

Z-test is used when the population standard deviation (œÉ) is known.

t-test is used when œÉ is unknown and estimated from the sample.

üìå This is the most important condition.

2. Large Sample Size

If the sample size is large (n ‚â• 30), a Z-test is appropriate.

With large samples, the sampling distribution of the mean is approximately normal (by the Central Limit Theorem).

3. Population is Normally Distributed (or approximately)

For small samples, the population should be normal.

For large samples, this condition is less strict.

4. Type of Data

Data should be quantitative (interval or ratio scale).

Observations must be independent.

Summary Table
Condition	Use Z-test	Use t-test
Population SD known	‚úÖ Yes	‚ùå No
Population SD unknown	‚ùå No	‚úÖ Yes
Sample size	Large (n ‚â• 30)	Small (n < 30)
Distribution	Normal / approx. normal	Normal (especially for small n)
Example

Z-test: A factory knows the population SD of bulb life and tests the mean life using a sample of 50 bulbs.

t-test: A researcher tests the average marks using a sample of 15 students where population SD is unknown.

One-Line Exam Answer

A Z-test is used when the population standard deviation is known and the sample size is large, whereas a t-test is used when the population standard deviation is unknown, especially for small samples.
